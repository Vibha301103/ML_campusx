```python
df1
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(df1['cgpa'],df1['package'])
plt.xlabel("cgpa")
plt.ylabel("package")
```

```python
x=df1.iloc[:,0:1]
y=df1.iloc[:,-1]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

X_test.iloc[0]

This selects the first row of your test features X_test.

The result is a pandas Series containing all feature values of that row.

.values

Converts the Series into a NumPy array.

Example: if the first row had one feature with value 5, .values gives array([5]).

.reshape(1,1)

Machine learning models in scikit-learn expect the input X to be 2D, shaped like (num_samples, num_features).

Here, you have 1 sample and 1 feature, so you reshape it to (1,1).

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

predicted value =3.8911
actual value=4.10
